## Building a backtesting system
Testing against last 100 days, but we want to test against the past years of data.

1. Create a prediction function
    2. Generating predictions
    3. Combining model into series
    4. Combining everything tgt
2. Backtest takes (data, ML model, predictors, start value(certain amt of data to train), step value(training every th year, eg take 11th year of data to predict the 12th year & so on.)
    2. create a list, list of df of each df is a prediction of a single year
    3. Create a loop function yr by yr to make predictions every year except the first 10 or so
    4.split training and test data where training is year prior to the current year and test is current year
    5. Use predict function to generate predictions
    6. Append to all predictions
    7. Concat, take a list of df to create into a single df

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index= test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

Backtest of our model w the sp500

```python
predictions = backtest(sp500, model, predictors)
```

## Rolling mean
horizons - [mean close price, last trading week, last 3 months, last year, last 4 years]
ratio between today's closing price and the closing in those period,
if mkt goes up a ton, due for a down turn vice versa

for each loop goes by the ^

trend finds the no of days when the stock price went up based on wtv horizon, shift forward to find look the back 4 days and sum it up to get avg


```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizons}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
    
    
```

```python
sp500 = sp500.dropna()
```

## Improving model
1. Rewrite predict function
    1.Model returns 0 or 1 currently
    2. Control over how we define 0 or 1
    3. predict_proba --> probability if it goes up or down
    4. set custom threshold, if made higher than 60% model becomes more confident that shows the price act goes up
        --> reduce the # of trading days, that predicts the price that goe up
        -->increase the chance that the price that ACTUALLY goes up


```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] =1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index= test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
